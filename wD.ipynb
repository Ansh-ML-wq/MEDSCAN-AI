```python
df["Sex"]=df["Sex"].map({"Male":1,"Female":0})
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df["Alcohol Use"]=le.fit_transform(df["Alcohol Use"])
```

```python
df["Socioeconomic Status"]=le.fit_transform(df["Socioeconomic Status"])
df["Region"]=le.fit_transform(df["Region"])
```

```python
df = df.drop(columns = ["Sex","Free Copper in Blood Serum","Albumin","Alkaline Phosphatase (ALP)",
                         "Prothrombin Time / INR","Psychiatric Symptoms","Region","Socioeconomic Status","Alcohol Use","BMI"],axis = 1)
```

```python
df.describe()
```

```python
x = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(x,y,test_size=0.33, random_state=42)
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
```

```python
models = []
models.append(("LogisticRegression",LogisticRegression()))
models.append(("RandomForestClassifier",RandomForestClassifier()))
models.append(("SVM",SVC()))
models.append(("Naive_baiyes",GaussianNB()))
models.append(("Tree",DecisionTreeClassifier()))
```

```python
scoring = "accuracy"
result = []
names = []
for name,model in models:
    kfold = KFold(n_splits=10, random_state=0,shuffle=True)
    cv_result = cross_val_score(model,X_train,y_train,cv=kfold,scoring=scoring)
    result.append(cv_result)
    names.append(name)
    print(model)
    print("%s %f"%(name,cv_result.mean()))
```

```python
clf = RandomForestClassifier()
clf.fit(X_train,y_train)
pred = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score
score = accuracy_score(y_test,pred)
p_s = precision_score(y_test,pred)
cf = confusion_matrix(y_test,pred)
```

```python
print(f"accuracy_score:{score}\n Precision_score:{p_s}")
```
